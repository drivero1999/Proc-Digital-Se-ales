```python
#Importar librerias
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

**Punto 1 =============================================================================**

```python
#Continua
def esc_unitario(t0,title, label_y):
  t = np.linspace(-5, 5, 1000)
  ut=np.piecewise(t, [t < t0, t > t0], [0, 1])
  plt.plot(t,ut)
  plt.title(title)
  plt.xlabel('t')
  plt.ylabel(label_y)
  plt.ylim(-1,1.5)
  plt.grid()
  plt.savefig("graph.pdf")
  plt.show()

#Discreta
def esc_unitario2(k,title, label_y):
  n = np.linspace(-5, 5, 40)
  un=np.piecewise(n, [n < k, n >= k], [0, 1])
  plt.stem(n, un,use_line_collection=True)
  plt.title(title)
  plt.xlabel('t')
  plt.ylabel(label_y)
  plt.ylim(-1,1.5)
  plt.grid()
  plt.savefig("graph.pdf")
  plt.show()
```

**u(t)**

```python

esc_unitario(0,"Señal u(t): Escalon Unitario", "u(t)")
```

**Punto 2 =============================================================================**

**Senoidal**

```python
def senoidal(A,f,desfase):
  t = np.linspace(0,5, 1000)
  w0=2*np.pi*f
  ut = A*np.sin(t*w0+desfase)
  plt.plot(t,ut)
  plt.title("Señal Senoidal")
  plt.xlabel('t')
  plt.ylabel('sen(t)')
  plt.grid()
  plt.savefig("senoCon.pdf")
  plt.show()

senoidal(3,1,np.pi/2)
```

**Cuadrada**

```python
def cuadrada(A,f,desfase):
  t = np.linspace(0,5, 1000)
  w0=2*np.pi*f
  ut = A*np.sign(np.sin(t*w0+desfase))
  plt.plot(t,ut)
  plt.title("Señal Cuadrada")
  plt.xlabel('t')
  plt.ylabel('sign(t)')
  plt.grid()
  plt.savefig("cuadraCon.pdf")
  plt.show()

cuadrada(3,1,np.pi/2)
```

**Diente de Sierra**

```python
def sawtooth(A,f,desfase):
  t = np.linspace(0,5, 1000)
  w0=2*np.pi*f
  ut = A * signal.sawtooth(t*w0+desfase)
  plt.plot(t,ut)
  plt.title("Señal Diente de Sierra")
  plt.xlabel('t')
  plt.ylabel('sawtooth(t)')
  plt.grid()
  plt.savefig("sawCon.pdf")
  plt.show()

sawtooth(3,1,np.pi/2)
```

**x(t)**

```python
def func_(x):
  p1,p2= (0,0),(4,3)
  m=(p2[1]-p1[1])/(p2[0]-p1[0])
  if(x<0 or x>4):
    return 0
  if(0<=x<=4):
    y=m*(x-p1[0])+p1[1]
    return y

def signal():
    t = np.linspace(-5, 7,1000)
    ut=[func_(t[i]) for i in range(len(t))]
    plt.plot(t,ut)
    plt.ylim(0,5)
    plt.title("Grafica x(t)")
    plt.xlabel('t')
    plt.ylabel('x(t)')
    plt.grid()
    plt.savefig("x(t).pdf")
    plt.show()

signal()
```

```python
def SAR_ADC(Uin, Uref):
  bits=8
  ans=[]
  cnt=0
  n=2
  for i in range(bits):
    aux=(1/n)*Uref
    if(cnt+aux <= Uin):
      ans.append(1)
      cnt+=aux
    else:
      ans.append(0)
    n=n*2
  listToStr = ' '.join(map(str, ans))
  print(listToStr)

SAR_ADC(0.1,7.4)
```
