```python
#Importar librerias
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

**Punto 1 =============================================================================**

```python
#Grafica Señal Cuadrada
Fs=1000
T=1/Fs
t = np.arange(0,1, T)
xt= signal.square(2 * np.pi * 5 * t)
plt.plot(t,xt)
plt.title("Señal Cuadrada x(t)")
plt.xlabel('tiempo (seg)')
plt.ylabel('x(t)')
plt.grid()
plt.ylim(-1.2,1.2)
plt.xlim(0,1)
plt.savefig("cuadra.pdf")
plt.show()
```

**Punto 2 =============================================================================**

```python
L=len(t)
XW=np.fft.fft(xt)/L
XW=np.fft.fftshift(XW)
frq=np.arange(-L/2,L/2,1) * (Fs/L)
plt.plot(frq, 2*abs(XW))
plt.title("Espectro de Frecuencia X(W)")
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('X(W)')
plt.grid()
plt.ylim(0,1.5)
plt.xlim(0,400)
plt.savefig("cuadraFourier2.pdf")
plt.show()
```

**Punto 3 =============================================================================**

```python
#Coeficientes
def reconst_fourier(numArmonicos):
  #Obtener coeficientes bk
  i=505
  bk=[(0,0)]
  while(len(bk)<numArmonicos+1):
    bk.append((i-500,2*abs(XW[i])))
    i+=5

  #Serie de Fourier
  t = np.linspace(0,1, 1000)
  sum=0
  k=1
  w=2*np.pi*5
  while(k<=numArmonicos):
    sum+=bk[k][1]*np.sin(k*w*t)
    k+=1

  plt.plot(t,sum)
  plt.title("Aproximacion Señal Cuadrada x(t)")
  plt.xlabel('tiempo (seg)')
  plt.ylabel('x(t)')
  plt.grid()
  plt.ylim(-1.5,1.5)
  plt.xlim(0,1)
  plt.savefig("serieCuad.pdf")
  plt.show()
```

```python
reconst_fourier(1)
```

```python
reconst_fourier(5)
```

```python
reconst_fourier(20)
```

```python
reconst_fourier(90)
```

**Punto 4 =============================================================================**

```python
#Obtener coeficientes bk
i=505
bk=[(0,0)]
while(len(bk)<10+1):
  bk.append((i-500,2*abs(XW[i])))
  i+=5
#print(bk[1:])

fig, axs = plt.subplots(5)
fig.suptitle('Primeros 5 armonicos impares')
axs[0].plot(t, bk[1][1]*np.sin(2*np.pi*bk[1][0]*t))
axs[1].plot(t, bk[3][1]*np.sin(2*np.pi*bk[3][0]*t))
axs[2].plot(t, bk[5][1]*np.sin(2*np.pi*bk[5][0]*t))
axs[3].plot(t, bk[7][1]*np.sin(2*np.pi*bk[7][0]*t))

axs[4].plot(t, bk[9][1]*np.sin(2*np.pi*bk[9][0]*t))

for ax in axs.flat:
    ax.set(xlabel='tiempo (seg)', ylabel='x(t)')
    ax.grid()

fig.savefig("invercuad.pdf")
```

```python
#Grafica Señal Diente de Sierra
Fs=1000
T=1/Fs
t = np.arange(0,1, T)
xt= signal.sawtooth(2 * np.pi * 5 * t)
plt.plot(t,xt)
plt.title("Señal Diente de Sierra x(t)")
plt.xlabel('tiempo (seg)')
plt.ylabel('x(t)')
plt.grid()
plt.ylim(-1.2,1.2)
plt.xlim(0,1)
plt.savefig("saw.pdf")
plt.show()
```

```python
#Transformada
L=len(t)
XW=np.fft.fft(xt)/L
XW=np.fft.fftshift(XW)
frq=np.arange(-L/2,L/2,1) * (Fs/L)
plt.plot(frq, 2*abs(XW))
plt.title("Espectro de Frecuencia X(W)")
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('X(W)')
plt.grid()
plt.ylim(0,1.5)
plt.xlim(0,400)
plt.savefig("sawF2.pdf")
plt.show()
```
